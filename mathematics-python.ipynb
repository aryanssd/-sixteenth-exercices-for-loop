```python
p = 0.2  # Change this to your desired probability


# Generate random samples from your bernoulli distribution 

num_samples = 50
samples = np.random.binomial(1, p, num_samples)

# Print the generated samples 
print(samples)
```

```python
# Get unique elements and their counts
unique_elements, counts = np.unique(samples, return_counts=True)

#Calculate percentages 
percentages = (counts / num_samples) * 100

#Create a dictionary to store the value counts and percentages
value_counts_and_percenteges = zip(unique_elements, counts, percentages)

# Print the value counts and percentages 

for i, j, k, in value_counts_and_percenteges : 
    print(f"{i}: Count: {j}, Percentage: {k:.2f}%")
```

```python
# Instence of Bernoulli distribution with paramete p = 0.8
bd = bernoulli(0.2)

# Outcome of random variable either 0 or 1 
x = [0,1]

# For the visualization of the bar plot of Bernoulli's distrubution \

fig, axs = plt.subplots(1, 2, figsize=(8, 3))

axs[0].bar(unique_elements, percentages, color="blue")
axs[0].set_xlim(-1, 2)
axs[0].set_ylim(0, 100)
axs[0].set_title('Bernoulli distribution (p=0.2 -sample', fontsize='10')
axs[0].set_xlabel('Values of random variable x (0, 1)', fontsize='10')
axs[0].set_ylabel('percenteges', fontsize='10') 



# Create the second bar plot


axs[1].bar(x, bd.pmf(x), color="green")
axs[1].set_xlim(-1, 2)
axs[1].set_ylim(0, 1)
axs[1].set_title('Bernoulli disribution (p=0.2) - Population' , fontsize = '10')
axs[1].set_xlabel('Values of random variable x (0, 1)' , fontsize='10')
axs[1].set_ylabel('Probability', fontsize='10')


# Adjust spacing between subplots

plt.tight_layout()

#Show the plots

plt.show()
```

```python
n = 5
p = 0.4 

num_samples = 100

samples = np.random.binomial(n, p, num_samples)

print(samples)

unique_elements, counts = np.unique(samples, return_counts = True)

percentages = (counts / num_samples) * 100 

value_counts_and_percentages = zip(unique_elements, counts, percentages)

for i, j, k, in value_counts_and_percentages:
    print(f"{i}: Count: {j}, Percentage: {k:.2f}%")
    
```

```python
bid = binom(n, p)

x = np.arange(0, n+1)

fig, axs = plt.subplots(1,2, figsize=(8,3))


axs[0].bar(unique_elements, percentages, color="blue")
axs[0].set_xlim(-1, n+1)
axs[0].set_ylim(0, 100)
axs[0].set_title('Binomial distribution (n,p) -sample', fontsize='10')
axs[0].set_xlabel('Values of random variable x', fontsize='10')
axs[0].set_ylabel('percenteges', fontsize='10') 

axs[1].bar(x, bid.pmf(x), color="green")
axs[1].set_xlim(-1, n+1)
axs[1].set_ylim(0, 1)
axs[1].set_title('Binomial disribution (n,p) - Population' , fontsize = '10')
axs[1].set_xlabel('Values of random variable x' , fontsize='10')
axs[1].set_ylabel('Probability', fontsize='10')



plt.tight_layout()

plt.show()
```

```python

y = binom.pmf (x, n, p)

print("Probability Mass Function of variable x")

for i, j in zip(x, y):
    print(f"{i}: {j:.2f}")

z = binom.cdf(x, n, p)

plt.title("Cumulative Distribution Function of variable x" , fontsize='10')
plt.xlabel("Values of variable x", fontsize= '10')
plt.ylabel("Cumulative Probability" , fontsize = '10')
plt.plot(x, z)
plt.show()
```
